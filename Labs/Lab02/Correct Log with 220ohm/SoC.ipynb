```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('log - copy.csv')

# Calculate State of Charge (SoC)
df['SoC'] = 100 - ((df['Capacity'] * 1) / 650)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['SoC'], df['Voltage1'], marker='o', linestyle='-', color='b', label='Voltage V1')
plt.title('Voltage V1 vs SoC')
plt.xlabel('State of Charge (SoC)')
plt.ylabel('Voltage V1')
plt.legend()
plt.gca().invert_xaxis()  # Reverse the x-axis scale
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Read the CSV file
df = pd.read_csv('log - copy.csv')

# Calculate State of Charge (SoC)
df['SoC'] = ((df['Capacity']*1) / 650)

# Apply Savitzky-Golay filter to smooth the 'Capacity' data
window_size_capacity = 5  # Adjust the window size as needed
order = 2  # Polynomial order, adjust as needed

df['Smoothed_Capacity'] = savgol_filter(df['Capacity'], window_size_capacity, order)

# Plot the data with smoothed capacity using Savitzky-Golay filter
plt.figure(figsize=(10, 6))
plt.plot(df['SoC'], df['Voltage1'], marker='o', linestyle='-', color='b', label='Voltage V1')
plt.plot(df['SoC'], df['Smoothed_Capacity'], color='r', label='Smoothed Capacity', linewidth=2)
plt.title('Voltage V1 vs SoC with Smoothed Capacity (Savitzky-Golay)')
plt.xlabel('State of Charge (SoC)')
plt.ylabel('Voltage V1')
plt.legend()
plt.grid(True)
plt.show()
```
