```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('log.csv')

# Calculate State of Charge (SoC)
df['SoC'] = df['Capacity'] / 54438072.97


# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['SoC'], df['Voltage1'], marker='o', linestyle='-', color='b', label='Voltage V1')
plt.title('Voltage V1 vs SoC')
plt.xlabel('State of Charge (SoC)')
plt.ylabel('Voltage V1')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create a scatter plot with a color gradient based on SoC
plt.scatter(df['SoC'], df['Voltage1'], c=df['SoC'], cmap='viridis', marker='o', label='Voltage V1')

plt.title('Voltage V1 vs SoC with Color Gradient')
plt.xlabel('State of Charge (SoC)')
plt.ylabel('Voltage V1')
plt.colorbar(label='State of Charge')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create an area plot
plt.fill_between(df['SoC'], df['Voltage1'], alpha=0.5, label='Voltage V1', color='skyblue')

plt.title('Voltage V1 vs SoC (Area Plot)')
plt.xlabel('State of Charge (SoC)')
plt.ylabel('Voltage V1')
plt.legend()
plt.grid(True)
plt.show()
```
