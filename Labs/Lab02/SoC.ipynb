```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = 'log.txt'  # replace with the actual path to your log.txt file
with open(file_path, 'r') as file:
    data = [line.strip().split(',') for line in file]
```

```python
# Extract time and voltage values
time = [int(entry[0]) for entry in data]
voltage = [float(entry[1]) for entry in data]

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(voltage, time, label='Voltage', marker='o', linestyle='-')
plt.title('Voltage vs Time')
plt.xlabel('Voltage')
plt.ylabel('Time')
plt.grid(True)
plt.legend()
plt.show()
```

```python

from mpl_toolkits.mplot3d import Axes3D

# Assuming you already have the 'time' and 'voltage' lists
# Extract time and voltage values
time = [int(entry[0]) for entry in data]
voltage = [float(entry[1]) for entry in data]

# Plot the 3D graph
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(voltage, time, [0] * len(time), label='Voltage', marker='o', linestyle='-')

# Set labels and title
ax.set_xlabel('Voltage')
ax.set_ylabel('Time')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot: Voltage vs Time')

# Show the plot
plt.legend()
plt.show()
```

```python
# Assuming you already have the 'time' and 'voltage' lists
# Extract time and voltage values
time = [int(entry[0]) for entry in data]
voltage = [float(entry[1]) for entry in data]

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(voltage, time, width=0.1, color='blue', edgecolor='black')

# Set labels and title
plt.title('Bar Graph: Time vs Voltage')
plt.xlabel('Voltage')
plt.ylabel('Time')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import numpy as np

# Assuming you already have the 'time' and 'voltage' lists
# Extract time and voltage values
time = [int(entry[0]) for entry in data]
voltage = [float(entry[1]) for entry in data]

# Create a scatter plot with color gradient based on voltage
plt.figure(figsize=(10, 6))
scatter = plt.scatter(time, np.zeros_like(time), c=voltage, cmap='viridis', marker='o', s=50, edgecolors='k')

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Voltage')

# Set labels and title
plt.title('Scatter Plot: Voltage vs Time')
plt.xlabel('Time')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```
