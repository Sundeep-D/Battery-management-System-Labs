```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import timedelta

# Read the CSV file
df = pd.read_csv('log.csv')

# Calculate State of Charge (SoC)
df['SoC'] = (1-((271-(df['TotalCapacity'])) / 271))*100

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['SoC'], df['Voltage1'], marker='o', linestyle='-', color='b', label='Voltage V1')
plt.title('Voltage vs SoC')
plt.xlabel('1 - State of Charge (%)')
plt.ylabel('Voltage V1')
plt.legend()
# plt.gca().invert_xaxis()  # Reverse the x-axis scale
plt.grid(True)
plt.show()

# Calculate time taken for full discharge
last_timestamp = df['Timestamp'].iloc[-1]  # Get the last timestamp
time_taken = timedelta(milliseconds=int(last_timestamp))
days, hours, minutes, seconds = time_taken.days, time_taken.seconds // 3600, (time_taken.seconds // 60) % 60, time_taken.seconds % 60

# Print the result
print(f"Time taken for full discharge: {days} days, {hours} hours, {minutes} minutes, {seconds} seconds")
```
